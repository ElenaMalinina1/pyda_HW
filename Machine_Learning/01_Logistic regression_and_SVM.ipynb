# Логистическая регрессия

```python
import numpy as np
import pandas as pd
import scipy.stats
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split 
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
```

```python
df = pd.read_csv('adult.csv')
df.head()
```

```python
df.info()
```

```python
# целевая переменная является категориальной

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(df['income'])
```

```python
# исходный тип данных
print(le.classes_)

# пример "расшировки" столбца income
print(le.transform(['<=50K', '>50K' ]))
```

```python
y = pd.Series(le.transform(df['income']))
y.head()
```

```python
# смотрим пустые значения
selectedColumns = df[['age','education','hours-per-week']]
selectedColumns.isnull()
```

```python
# столбец education - категориальная переменная 
# переведем ее в значения 0 и 1, добавив столбцы с соответствующими названиями
X = pd.get_dummies(selectedColumns, columns = ['education'])
X
```

```python
model = LogisticRegression()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

model = make_pipeline(StandardScaler(), 
                      LogisticRegression(max_iter=1000))
```

```python
model.fit(X_train, y_train)
predictions = model.predict(X_test)
```

```python
model.score(X_train, y_train) 
```

```python
model.score(X_test,y_test)
```

## SVM

```python
clf = make_pipeline(StandardScaler(), SVC(gamma='auto')) 
clf.fit(X_train, y_train)
```

```python
clf.score(X_train, y_train) 
```

```python
clf.score(X_test, y_test)
```
