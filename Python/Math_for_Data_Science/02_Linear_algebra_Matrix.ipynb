# Уровень 1:
Дан объект в 2d пространстве.

При помощи матричных операций выполните:

    Уменьшить объект в два раза
    Повернуть на 130 градусов
    Отразить объект относительно прямой y=x
    Выполнить смещение объекта и масштабирование (применить 2 матрицы трансформаций). Смещение на (200, 300) и масштабирование на (0.5, 1.2)

```python
A = np.array([
    [-100, -100, 1],
    [-100,  100, 1],
    [100,   100, 1],
    [0,     0,   1],
    [100,  -100, 1],
    [-100, -100, 1]
])

x = A[:,0]
y = A[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.title('Изначальный объект')
plt.grid()
plt.show()
```

```python
A_small = A @ np.array([[0.5, 0,   0],
                        [0,   0.5, 0],
                        [0,   0,   1]])

x = A_small[:,0]
y = A_small[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.title('Уменьшенный в 2 раза объект')
plt.grid()
plt.show()
```

```python
rad_130deg = np.deg2rad(130)


A_rotate = A @ np.array([[np.cos(rad_130deg), -np.sin(rad_130deg), 0], 
                         [np.sin(rad_130deg), np.cos(rad_130deg),  0],
                         [0,                  0,                   1]])

x = A_rotate[:,0]
y = A_rotate[:,1]

plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200])
plt.title('Повернутый на 130 градусов объект')
plt.grid()
plt.show()
```

```python
A_y_eq_x = (A @ np.array([[0,   1,   0],
                          [1,   0,   0],
                          [0,   0,   1]]))
x = A_y_eq_x[:,0]
y = A_y_eq_x[:,1]

plt.plot(x, y)
plt.ylim([-200, 600])
plt.xlim([-200, 600])
plt.title('Объект в представлении y=x')
plt.grid()
plt.show()
```

```python
A_scale = (A @ np.array([[1,   0,   0],
                         [0,   1,   0],
                         [200, 300, 1]])) @ np.array([[0.5, 0,   0],
                                                      [0,   1.2, 0],
                                                      [0,   0,   1]])
x = A_scale[:,0]
y = A_scale[:,1]

plt.plot(x, y)
plt.ylim([-200, 600])
plt.xlim([-200, 600])
plt.title('Объект с перемещением на {200, 300] и масштабированием [0.5, 1.2]')
plt.grid()
plt.show()
```
