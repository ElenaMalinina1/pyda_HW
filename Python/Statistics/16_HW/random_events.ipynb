# Задание
Смоделировать игру против лотерейного автомата типа “777”. Игрок платит 1 руб., после чего выпадает случайное целое число, равномерно распределенное от 0 до 999. При некоторых значениях числа игрок получает выигрыш.

    777: 200 руб.
    999: 100 руб.
    555: 50 руб.
    333: 15 руб.
    111: 10 руб.
    *77: 5 руб.
    **7: 3 руб.
    *00: 2 руб.
    **0: 1 руб.

Вопросы:

    Выгодна ли игра игроку?
    Сколько в среднем приобретает или теряет игрок за одну игру?
    
* Дополнительное задание повышенной сложности.

Теоретически рассчитать средний выигрыш (проигрыш) и сравнить с результатами моделирования

– любая цифра

```python
import numpy as np
```

```python
trials = 1000000    # Количество попыток
sum_per_trial= 1 # Цена 1 игры на автомате,рублей
sum_bets =  trials*sum_per_trial   # Всего денего потрачено на ставки 
successes =0     # Количество выйгрышей
sum_wins = 0    # Сумма выйгрыша в рублях
lucky_numbers={ 
    777:200,
    999:100,
    555:50,
    333:15,
    111:10
     } # Выигрышные номера и сумма выйгрыша в рублях, часть1
for i in range(trials):
    random_score = np.random.randint(0,1000)    # Выпавшее число от 0 до 999
    if random_score in lucky_numbers:
        successes += 1
        sum_wins += lucky_numbers[random_score]
    elif random_score%100 == 77:
        successes += 1
        sum_wins += 5
    elif random_score%10 == 7:
        successes += 1
        sum_wins += 3 
    elif random_score%100 == 0:
        successes += 1
        sum_wins += 2
    elif random_score%10 == 0:
        successes += 1
        sum_wins += 1
        
p_monte_carlo = successes / trials    # Вероятность из моделирования

#Расчитаем вероятность выйгрыша
# 5 выйгрышных комбинаций из списка lucky_numbers + 9(*77)+90(**7)+9(*00)+90(**0)=203
p = 203 / 10000  # Теоретическая вероятность
#Расчитаем среднюю сумму выйгрыша
x = 200/1000+100/1000+50/1000+15/1000+10/1000+5*9/1000+3*90/1000+2*9/1000+1*90/1000
print('Данные эксперимента:')
print('Попыток:',trials)
print('Потрачено денег на ставки,рублей :',sum_bets)
print('Количество выигрышей, раз:',successes)
print('Сумма выигрышей,рублей:',sum_wins)
print('Баланс после игры,рублей:',sum_wins-sum_bets,'\n')
print('Расчетные данные:')
print('Расчетный выигрыш,рублей:',x*trials)
print('Расчетный баланс после игры,рублей:',x*trials-trials)
print('Средний выигрыш(+) или проигрыш(-) за игру,рублей:',x-sum_per_trial,'\n')
print('Теоретическая вероятность выигрыша: ',p)
print('Вероятность выигрыша из моделирования по методу Монте-Карло: ',p_monte_carlo,'\n')
if sum_wins-sum_bets < 0:
    print('Игра игроку не выгодна!')
else:
    print('Игра игроку выгодна!')
```
