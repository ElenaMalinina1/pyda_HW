столбец Age. Максимальное значение 9, а по описанию 2. Выбросы есть

столбец pulse. Максимальное значение 184, а по описанию 20-40. Выбросы есть

столбец respiratory rate. Максимальное значение 96, а по описанию 8-10 (хотя в описании написано, что полезность столбца сомнительна, из-за больших отклонений) Все остальные столбцы в содержат значения в допустимой норме, исходя из описания.

1: surgery? 1 = Yes, it had surgery 2 = It was treated without surgery

2: Age 1 = Adult horse 2 = Young (< 6 months)

4: rectal temperature - linear - in degrees celsius. - An elevated temp may occur due to infection. - temperature may be reduced when the animal is in late shock - normal temp is 37.8 - this parameter will usually change as the problem progresses eg. may start out normal, then become elevated because of the lesion, passing back through the normal range as the horse goes into shock 

5: pulse - linear - the heart rate in beats per minute - is a reflection of the heart condition: 30 -40 is normal for adults - rare to have a lower than normal rate although athletic horses may have a rate of 20-25 - animals with painful lesions or suffering from circulatory shock may have an elevated heart rate

6: respiratory rate - linear - normal rate is 8 to 10 - usefulness is doubtful due to the great fluctuations

7: temperature of extremities - a subjective indication of peripheral circulation - possible values: 1 = Normal 2 = Warm 3 = Cool 4 = Cold - cool to cold extremities indicate possible shock - hot extremities should correlate with an elevated rectal temp. 

11: pain - a subjective judgement of the horse's pain level - possible values: 1 = alert, no pain 2 = depressed 3 = intermittent mild pain 4 = intermittent severe pain 5 = continuous severe pain - should NOT be treated as a ordered or discrete variable! - In general, the more painful, the more likely it is to require surgery - prior treatment of pain may mask the pain level to some extent 

23: outcome - what eventually happened to the horse? - possible values: 1 = lived 2 = died 3 = was euthanized

### Задание 2. Первичное изучение данных

```python
# поиск межквартильного размаха по пульсу
q1 = sick_horse['pulse'].quantile(0.25)
q3 = sick_horse['pulse'].quantile(0.75)
IQR = q3 - q1
lower_bound = q1 - (1.5 * IQR) 
upper_bound = q3 + (1.5 * IQR)
remove_outliers = sick_horse[sick_horse['pulse'].between(lower_bound, upper_bound, inclusive=True)]
# вывод выбросов
pd.concat([sick_horse, remove_outliers]).drop_duplicates(keep=False).sort_values('pulse')
# в норме 30 - 40 ударов в минуту, у этих лошадей пульс значительно превышен, возможно они больны
```

```python
# поиск межквартильного размаха по частоте дыхания
q1 = sick_horse['respiratory rate'].quantile(0.25)
q3 = sick_horse['respiratory rate'].quantile(0.75)
IQR = q3 - q1
lower_bound = q1 - (1.5 * IQR) 
upper_bound = q3 + (1.5 * IQR)
remove_outliers = sick_horse[sick_horse['respiratory rate'].between(lower_bound, upper_bound, inclusive=True)]
# вывод выбросов
pd.concat([sick_horse, remove_outliers]).drop_duplicates(keep=False).sort_values('respiratory rate')
# в норме 8-16 вдохов и выдохов в минуту, лошадкам явно нехорошо 

# оставляю выбросы и продолжу работу с ними т.к. судя по пульсу, частоте дыхания - это естественные выбросы(лошади болееют)
```

### Задание 3. Работа с пропусками

```python
# общее кол-во пропусков %
(sick_horse.isna().mean() * 100).round(2)
```

```python
# т.к у нас отсутствуют столбцы с пропусками более 80%, то все столбцы оставляем
```

```python
print(sick_horse['temperature of extremities'].value_counts())
print('-' * 30)
print(sick_horse['pain'].value_counts())
print('-' * 30)
print(sick_horse['outcome'].value_counts())
```

```python
# Гипотеза - проверим взаимосвязь между столбцами pain и outcome с группировкой по возрасту лошади
print(sick_horse.groupby(['age','pain'])['outcome'].agg(pd.Series.mode))
```

```python
# можно сделать вывод, что молодые лошади при коликах практически всегда погибают, (исключение тольуо уровень боли 3).
# Взрослые лошади при низком уровне боли (до 4) остются живыми, а с 4 уровня боли погибают все лошади. 
# Признаем эту гипотезу допустимой. Поэтому недостающие значения в столбце outcome заполняеем исходя из значения столбца pain
```

```python
#Сначала заполним недостающими значениями столбец outcome (у него самый маленький процент пропусков)
#Все действия будем производить с новым датафреймом, для быстрого отката в случае неправильного применения фильтра
#Заполним недостающие значениями столбца outcome

fill_horses_df = sick_horse
fill_horses_df['outcome'].fillna(fill_horses_df.groupby(['pain','age'])['outcome'].transform('median'), inplace=True)
fill_horses_df.describe()
```

```python
# Гипотеза - проверим взаимосвязь между столбцами surgery?(операция) и pain (боль) (тем более это описано в документации)
print(sick_horse.groupby(['surgery'])['pain'].median())
```

```python
# взаимосвязь есть - операция была (1) при высоком медианном значении боли(3), и не было(2) при низком значении боли(2)
```

```python
#заполним недостающие значениями столбца pain исходя из медианных значений столбца surgery
fill_horses_df['pain'].fillna(sick_horse.groupby(['surgery'])['pain'].transform('median'), inplace=True)
fill_horses_df.describe()
```

```python
#  в подтверждение гипотезы роверим обратную взаимосвязь между столбцами pain(боль) и surgery(хтрургическое вмешательство)
print(sick_horse.groupby(['pain'])['surgery'].median())
```

```python
# взаимосвязь есть - что на уровнях боли 3 и выше всегда было хирургическое вмешательство. Восстановим отсутствующее значение в столбце 'surgery'
```

```python
fill_horses_df['surgery'].fillna(fill_horses_df.groupby(['pain'])['surgery'].transform('median'), inplace=True)
fill_horses_df.describe()
```

```python
# Гипотеза - проверим взаимосвязь между outcome  и temperature of extremities
print(sick_horse.groupby(['outcome'])['temperature of extremities'].median())
```

```python
# Вывод - выявлена прямая связь-в случаях смертности (outcome =2 или 3),температура экскрементов (temperature of extremities) всегда была 3(это признак шока - из описания). 
# В случае если outcome = 1 (лошадь осталась живой), температура экскрементов = 2 (норма)
```

```python
#Восстановим отсутствующие значения в столбце 'temperature of extremities' 

fill_horses_df['temperature of extremities'].fillna(fill_horses_df.groupby(['outcome'])['temperature of extremities'].transform('median'), inplace=True)
fill_horses_df.describe()
```

```python
# Гипотеза- проверим взаимосвязь между столбцами temperature of extremities и pulse
print(sick_horse.groupby(['temperature of extremities'])['pulse'].median())
```

```python
# Вывод- выявлена прямая связь - чем ниже температура экскрементов(признак заболевания) - тем выше пульс
```

```python
#Восстановим отсутствующие значения в столбце 'pulse' 

fill_horses_df['pulse'].fillna(fill_horses_df.groupby(['temperature of extremities'])['pulse'].transform('median'), inplace=True)
fill_horses_df.describe()
```

```python
#Гипотеза - проверим утверждения из описания, что ректальная температура повышается во время инфекции,а во время болевого 
#шока(значение-pain) температура понижается
print(sick_horse.groupby(['pain'])['rectal temperature'].median())
```

```python
# Вывод- гипотезу можно признать успешной. При высоких показателях боли медианная температура понижается.
```

```python
#Восстановим отсутствующие значения в столбце 'rectal temperature' 

fill_horses_df['rectal temperature'].fillna(fill_horses_df.groupby(['pain'])['rectal temperature'].transform('median'), inplace=True)
fill_horses_df.describe()
```

```python
# Остался последний показатель- respiratory rate. В описании так и написано- полезность сомнительная из-за больших колебаний. Поэтому заполним недостающие значения медианными.
```
